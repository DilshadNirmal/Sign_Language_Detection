```python
import torch
import onnx
```

# load the pytorch model

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp21/weights/last.pt')
```

# Convert the model to ONNX format

```python
input_shape = (3, 640, 640)
x = torch.randn(1, *input_shape)
onnx_model_path = "model.onnx"
torch.onnx.export(model, x, onnx_model_path, opset_version=11, input_names=['input'], output_names=['output'], dynamic_axes={'input': {0: 'batch_size'}, 'output': {0: 'batch_size'}})
```
